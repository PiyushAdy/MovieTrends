```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('imdb.csv')
df.replace("\\N",np.nan,inplace=True)
df["startYear"]=pd.to_datetime(df["startYear"],errors='coerce').dt.year
df["endYear"]=pd.to_datetime(df["endYear"],errors='coerce').dt.year
df
```

```python
df['genres'] = df['genres'].str.replace("\\N","").str.split(',')
df_exploded = df.explode('genres')
df_exploded
```

Pivot Table for Number of Movies per year per genre

```python
genre_table = df_exploded.pivot_table(
    index='genres',           
    columns='startYear',      
    values='imdbID',         
    aggfunc='count'          
)
genre_table = genre_table.fillna(0)
genre_table
```

No if Movies Per genre Per year V2

```python
year_table=df_exploded.pivot_table(
    index='startYear',
    columns='genres',
    values='imdbID',
    aggfunc='count'
)
year_table = year_table.fillna(0)
year_table = year_table.reset_index()
# year_table.to_json(orient="index")
year_table.to_json(orient="records")
```

No of Movies Per Year

```python
yearwise_count = df.groupby('startYear')['imdbID'].count()
yearwise_count = yearwise_count.reset_index()  # Convert Series to DataFrame
yearwise_count.rename(columns={"imdbID": "Count"}, inplace=True)
yearwise_count.to_json(orient="records")
```

Average Runtime minutes per year

```python
df['runtimeMinutes'] = pd.to_numeric(df['runtimeMinutes'], errors='coerce')
df['runtimeMinutes'] = df['runtimeMinutes'].fillna(0)
yearly_runtime = df.groupby('startYear')['runtimeMinutes'].mean()
yearly_runtime = yearly_runtime.reset_index()
yearly_runtime.rename(columns={"runtimeMinutes": "Average_Runtime"}, inplace=True)
yearly_runtime.to_json(orient="records")
# yearly_runtime.plot(x='startYear', y='Average_Runtime', kind='line', marker='o', title='Average Runtime VS Year', grid=True)
# plt.xticks(rotation=45)
# plt.show()
```
